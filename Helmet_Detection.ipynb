# II. Train YOLOv10 for Helmet Detection task

**1. Download Helmet Safety dataset**

```python
%cd content
!gdown '1twdtZEfcw4ghSZIiPDypJurZnNXzMO7R'
!mkdir safety_helmet_dataset
!unzip -q '/content/Safety_Helmet_Dataset.zip' -d '/content/safety_helmet_dataset'
```

**2. Initialize YOLOv10 model**

```python
%cd yolov10
from ultralytics import YOLOv10

MODEL_PATH = 'yolov10n.pt'
model = YOLOv10(MODEL_PATH)
```

**3. Training**

```python
#Maximum trainable batch_size = 32

YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 50
IMG_SIZE = 640
BATCH_SIZE = 64

model.train (data = YAML_PATH,
    epochs = EPOCHS,
    batch = BATCH_SIZE,
    imgsz = IMG_SIZE)
```

**4. Evaluate**

```python
TRAINED_MODEL_PATH = 'runs/detect/train2/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
```

**5. Run prediction with an image**

```python
# https://github.com/googlecolab/colabtools/issues/3409
import locale
locale.getpreferredencoding = lambda: "UTF-8"
```

```python
from google.colab.patches import cv2_imshow

TRAINED_MODEL_PATH = 'runs/detect/train2/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

IMAGE_URL = 'https://ips-dc.org/wp-content/uploads/2022/05/Black-Workers-Need-a-Bill-of-Rights.jpeg'
CONF_THRESHOLD = 0.3
results = model.predict(source=IMAGE_URL,
                       imgsz=IMG_SIZE,
                       conf=CONF_THRESHOLD)
annotated_img = results[0].plot()

cv2_imshow(annotated_img)
```
